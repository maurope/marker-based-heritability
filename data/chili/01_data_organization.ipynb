```python
import pandas as pd
```

# Data input

### Cayenne

```python
c = pd.read_csv('cayenne.csv', skiprows=1, header = None)
c['variety'] = 'cayenne'
c.head(2)
```

### Tabasco

```python
t = pd.read_csv('tabasco.csv', skiprows=1, header = None)
t['variety'] = 'tabasco'
t.head(2)
```

### Habanero

```python
h = pd.read_csv('habanero.csv', skiprows=1, header = None)
h['variety'] = 'habanero'
h.head(2)
```

```python
df = pd.concat([c, t, h])
df.reset_index(drop=True, inplace=True)
df
```

```python
df.columns = ["Ng10.a", "Ng10.b", "Ng18.a", "Ng18.b","Ng8.a", "Ng8.b", "Bd12.a", "Bd12.b", "Ng20.a", "Ng20.b", "Ng6.a", "Ng6.b", "Ng7.a", "Ng7.b", "stem_width", "stem_height", "first_flower_day", "fruits_number", "fruit_length", "fruit_diameter", "fruit_weight","seeds_number", "yield", "variety"]
```

```python
df
```

```python
df["individual"] = [f"ind_{i}" for i in range(1, len(df) + 1)]
df = df[["individual"] + [col for col in df.columns if col != "individual"]]
```
